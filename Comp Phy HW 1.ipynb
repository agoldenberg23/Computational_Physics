```python
#Problem 1
#Familiarize yourself enough with a plotting package in Python (e.g. matplotlib) to plot
#a Gaussian with zero mean and a standard deviation of 5 over the range [-15,15]. Make
#sure that the Gaussian is normalized correctly.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 0
std = 5
#variance = np.square(std)
#print("mean:" + str(mean) + " std:" + str(std) + " var:" + str(variance))
x = np.arange(-15, 15, 0.00001)
y = norm.pdf(x, mean, std)

plt.plot(x,y)
plt.title("Gaussian Distribution")
plt.xlabel("X")
plt.ylabel("Probability")
plt.show()
```

```python
#Problem 3
#Exercise (2) in S2.4.3 of Landau

import numpy as np

#Part A
print("For double precision floats:")
under = np.float64(1.0)
over = np.float64(1.0)
#Number of itterations to run over
N = 1100
#Number to devide and multiply by
a = 2

for i in range(0,N,1):
    under = under / a
    over = over * a
    print("Iteration: " + str(i) + " under: " + str(under) + " over: " + str(over))


#Part B
print("\n\n\nFor single precision floats:")
under2 = np.float32(1.0)
over2 = np.float32(1.0)
#Number of itterations to run over
N = 1100
#Number to devide and multiply by
a = 2

for i in range(0,N,1):
    under2 = under2 / a
    over2 = over2 * a
    print("Interation: " + str(i) + " under: " + str(under2) + " over: " + str(over2))


#Part C
print("\n\n\nFor int (32):")
under3 = np.int32(-2147483000)
over3 = np.int32(2147483000)
#Number of itterations to run over
N = 1000
#Number to add and subtract by
a = 1

for i in range(0,N,1):
    under3 = under3 - a
    over3 = over3 + a
    print("Interation: " + str(i) + " under: " + str(under3) + " over: " + str(over3))
```

We can see the over and underflow from the output produced by the code.
For 64 bit floats, the overflow is at iteration 1023 when 8.98846567431158e+307 went to inf. The underflow is at iteration 1074 when 5e-324 went to 0.0.

For 32 bit floats, the overflow is at iteration 1023 when 8.98846567431158e+307 went to inf. The underflow is at iteration 1074 when 5e-324 went to 0.0. This means that numpy is saving the float as a 64 bit number despite me using np.float32() to declare a 32 bit number. From what I can gather from looking online, numpy does not store 32 bit floats on a 64 bit windows computer.

For the 32 bit int, the overflow is at iteration 647 when 2147483647 went to -2147483648. The underflow is at iteration 648 wen -2147483648 went to 2147483647.
