Part 1] Numerical Differentiation

The derivitive of f(x) exaluated at x is given by the finite formula.

Foward Diff: $f(x) = \dfrac{f(x + h) - f(x)}{h}$

Central Diff: f(x)= $\dfrac{f(x - \dfrac{h}{2}) - f(x + \dfrac{h}{2})}{h}$

We will apply this to the folowing function.

$f(x) = e^{x} sin(x)$ at x = 0.1, 1.0, 100

$\dfrac{d}{dx} f(x) = e^{x} (cos(x) + sin(x))$

```python
#Part A] For all 2 methods, print out the derivative and its relative error E as functions of step size h. Include 
#in the step size range the h-value where E equals machine precision E ≃ 10^−15 (use double-precision).

import math
import numpy as np
import matplotlib.pyplot as plt

#Forward diff formula
def forDiff(x, h):
    e = (math.e)**(x + h)
    s = math.sin(x + h)
    part = e * s
    e = (math.e)**(x)
    s = math.sin(x)
    part = part - (e * s)
    return abs(part / h)

#Central diff formula
def centDiff(x, h):
    e = (math.e)**(x - (h/2))
    s = math.sin(x - (h/2))
    part = e * s
    e = (math.e)**(x + (h/2))
    s = math.sin(x + (h/2))
    part = part - (e * s)
    return abs(part / h)

#Actual value for the derivative
def derF(x):
    e = (math.e)**x
    s = math.sin(x)
    c = math.cos(x)
    return e * (s + c)
```

```python
#The range of h values to be used
H = np.zeros(206, dtype="float64")
for i in range(6, 206):
    H[i] = 0.01 + H[i-1]
    
H[0] = 10**-17
H[1] = 10**-17
H[2] = 10**-16
H[3] = 10**-15
H[4] = 10**-14
H[5] = 10**-13

#For x = 0.1
resultfor_01 = np.zeros(np.size(H))
resultcent_01 = np.zeros(np.size(H))
errorfor_01 = np.zeros(np.size(H))
errorcent_01 = np.zeros(np.size(H))
exact = derF(0.1)

print("Exact = " + str(exact))

#Fill the arrays
for i in range(0, np.size(H)):
    h = H[i]
    resultfor_01[i] = forDiff(0.1, h)
    resultcent_01[i] = centDiff(0.1, h)
    errorfor_01[i] = ((exact - resultfor_01[i]) / exact)
    errorcent_01[i] = ((exact - resultcent_01[i]) / exact)

#Plot the result
plt.figure(1)
plt.title("Derivative of x = 0.1")
plt.xlabel("h")
plt.ylabel("Derivative")
plt.plot(H, resultfor_01, label="Forward Diff")
plt.plot(H, resultcent_01, label="Central Diff")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

plt.figure(2)
plt.title("Error of x = 0.1")
plt.xlabel("h")
plt.ylabel("Error")
plt.plot(H, errorfor_01, label="Forward Diff")
plt.plot(H, errorcent_01, label="Central Diff")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

We will implement the following algorithms from the textbook and one which is the actual second derivative.

$\dfrac{d^{2} f(x)}{d x^{2}} = \dfrac{[f(x + h) - f(x)] - [f(x) - f(x - h)]}{h^{2}}$, Equation: (5.22)

$\dfrac{d^{2} f(x)}{d x^{2}} = \dfrac{f(x + h) + f(x - h) - 2f(x)}{h^{2}}$, Equation: (5.23)

Actual second derivative: $\dfrac{d^{2} f(x)}{d x^{2}} = 2 e^{x} cos(x)$

```python
#Part C] Repeat part A for the second-derivative using the 2 central difference algorithms in
#Eq. 5.22 and 5.23 in Landau. Which is more accurate? Why?

import math

#Equation 5.22 (the 1st equation)
def secDer1(x, h):
    e = (math.e)**(x + h)
    s = math.sin(x + h)
    part = e * s
    e = (math.e)**(x)
    s = math.sin(x)
    part2 = e * s
    part = part - part2
    e = (math.e)**(x - h)
    s = math.sin(x - h)
    part2 = part2 - (e * s)
    part = part - part2
    return part / (h * h)

#Equation 5.23 (the 2nd equation)
def secDer2(x, h):
    e = (math.e)**(x + h)
    s = math.sin(x + h)
    part = e * s
    e = (math.e)**(x - h)
    s = math.sin(x - h)
    part = part + (e * s)
    e = (math.e)**(x)
    s = math.sin(x)
    part = part - (2 * (e * s))
    return part / (h * h)

#The actual value
def secDer(x):
    e = (math.e)**(x)
    c = math.cos(x)
    return 2 * e * c
```

```python
import numpy as np

#The range of h values to be used
H = np.zeros(206, dtype="float64")
for i in range(6, 206):
    H[i] = 0.01 + H[i-1]
    
H[0] = 10**-17
H[1] = 10**-17
H[2] = 10**-16
H[3] = 10**-15
H[4] = 10**-14
H[5] = 10**-13

#For x = 0.1
result1_01 = np.zeros(np.size(H))
result2_01 = np.zeros(np.size(H))
error1_01 = np.zeros(np.size(H))
error2_01 = np.zeros(np.size(H))
exact = secDer(0.1)

print("Exact = " + str(exact))

#Fill the arrays
for i in range(0, np.size(H)):
    h = H[i]
    result1_01[i] = secDer1(0.1, h)
    result2_01[i] = secDer2(0.1, h)
    error1_01[i] = ((exact - result1_01[i]) / exact)
    error2_01[i] = ((exact - result2_01[i]) / exact)
    
#Plot the result
plt.figure(9)
plt.title("Second Derivative of x = 0.1")
plt.xlabel("h")
plt.ylabel("Second Derivative")
plt.ylim(0, 4)
plt.xlim(0.1, 2)
plt.plot(H, result1_01, label="Eq. 5.22")
plt.plot(H, result2_01, label="Eq. 5.23")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

plt.figure(10)
plt.title("Error of x = 0.1")
plt.xlabel("h")
plt.ylabel("Error")
#plt.ylim(0, 0.5)
plt.xlim(0.1, 2)
plt.loglog(H, error1_01, label="Eq. 5.22")
plt.loglog(H, error2_01, label="Eq. 5.23")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

The cubic spline interpolation uses a different polynomial of third power for each iteration (space between each pair of data points) to approximate the values. This is similar to the method used in question 2 part C and D but differs in the conditions added to make the function differentiable. This is done by adding the requirements that the first and second derivatives at each of the data points (where two polynomials meet) must be equal. Thus, kinks and other non-differentiable aspects are not allowed. The data points at the end function only have one polynomial that touches the point, so we cannot set this equal. Two boundary conditions are needed after the conditions stated along with the condition of the interpolation going through all data points. Thus, we set the function ends second derivative to 0 to create two conditions.
